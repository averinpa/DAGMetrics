# Compare Markov blankets of selected nodes with DAGMetrics

```r
library(bnlearn)
library(DAGMetrics)
```

## Create random DAG

```r
set.seed(123)
nodes <- 50
prob <- 0.04
bn_1 <- random.graph(nodes = paste0("S", 1:nodes), method = "ordered", prob = prob)
bn_2 <- random.graph(nodes = paste0("S", 1:nodes), method = "ordered", prob = prob)
```

## Visualize Markov blankets of selected nodes

```r
vis_dag(bn_2, nodes = c('S3'))
```

```r
vis_dag(bn_2, nodes = c('S3', 'S49'))
```

## Compare Markov blankets of selected nodes

```r
vis_dag_comparison(bn_1, bn_2, nodes = c('S3'))
```

```r
vis_dag_comparison(bn_1, bn_2, nodes = c('S3', 'S49'))
```
